# 5-Like/Unlike posts

### 1-Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
def Search_dilsefoodie():
    # search 'dilsefoodie' using xpath
    search = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Search']")))
    # type in search bar
    search.send_keys("dilsefoodie")
    # take time for opening 'dilsefoodie' page
    driver.implicitly_wait(3)
    # hit enter using Keys
    enter = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//div[starts-with(@class,'_abnx')]/div")))
    enter.click()
    driver.implicitly_wait(5)
    
    enter_post = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, '_aabd _aa8k _aanf')]")))  
    enter_post.click() 
    time.sleep(2)
    
    count = 1
    while count <= 30:
        html = driver.find_element_by_xpath("//span[@class='_aamw']//span").get_attribute('outerHTML')
        like = data = BeautifulSoup(html, 'html.parser').svg['aria-label']
        if like == 'Like':
            driver.find_element_by_xpath("//span[@class='_aamw']/button").click()
            print('Like', count)
        else:
            print('Already Like', count)
        count += 1
        driver.implicitly_wait(10)
        driver.find_element_by_xpath("//div[contains(@class,' _aaqg _aaqh')]/button").click()
```

### Print 500 followers

```python
def Print500followers():
    # list of followers using xpath
    followers = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[starts-with(@class, '_ab8w')]/div/div/span/a")))
    # followers len is 12 (12 username)
    count = 1
    scroll = 0
    try:
        while scroll < 51:  # 12X42 = 504
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', followers)
            time.sleep(2)
            scroll += 1
            for i in followers:
                # get HTML data and find followers username
                html = i.get_attribute('innerHTML')
                data = BeautifulSoup(html, 'html.parser').string
                if data != "" and count <= 500:
                    print(count, data)
                    count += 1
            if count >= 500:
                break
    except Exception as e:
        print(e)
```

### 2-Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

### find the followers of “foodtalkindia” that you are following (Mutual Followers)

```python
def Mutual_followers():
    # close followers window
    driver.back()
    time.sleep(2)
    # find mutual followers usnig partial link text
    followersBy = driver.find_element_by_partial_link_text('Followed by').click()
    time.sleep(2)
    followers = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[starts-with(@class, ' _ab8y')]")))
    #print(len(followers))
    
    try:
        following = []
        while True:
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', followers)
            time.sleep(2)
            for i in followers:
                f = i.get_attribute('innerHTML').strip()
                if f != '.':
                    following.append(f)
                    #print(f)
            if len(following) == len(followers):
                break
    except Exception as e:
        print(e)
    return following
```

### find my followers

```python
def MyFollowers():
    # go to back page
    #driver.back()
    profile = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[starts-with(@class, '_aarf _aak0')]")))
    profile.click()
    time.sleep(2)
    
    followers = driver.find_element_by_partial_link_text(' followers').click()
    time.sleep(3)
    Myfollower = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[starts-with(@class, ' _ab8y  _ab94 _ab97 _ab9f _ab9k _ab9p _abcm')]")))
    
    try:
        MyFollwers = []
        while True:
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', Myfollower)
            time.sleep(2)
            for i in Myfollower:
                MyFollwers.append(i.get_attribute('innerHTML'))
                #print(i.get_attribute('innerHTML'))
            if len(MyFollwers) == len(Myfollower):
                break
    except Exception as e:
        print(e)
    return MyFollwers
```

### After view story run story_status

```python
def story_status():
    
    height = driver.find_element_by_xpath("//div[starts-with(@class,'_aarf')]//canvas")
    html = height.get_attribute('outerHTML')
    
    height = BeautifulSoup(html, 'html.parser').find('canvas')['height']
    width = BeautifulSoup(html, 'html.parser').find('canvas')['width']
    
    if int(height) == 166 and int(width) == 166:
        print("Already Viewed")
    if int(height) == 168 and int(width) == 168:
        enter = driver.find_element_by_xpath("//div[starts-with(@class,'_aarf')]").click()
        try :
            driver.find_element_by_xpath("//section[@contains(class,'carul')]")
        except NoSuchElementException:
            print("User has no story")
```
